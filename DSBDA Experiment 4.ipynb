```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_dataset=load_boston()
```

```python
print(boston_dataset.keys())
```

```python
data=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
data.head()
```

```python
#Add target value price which is missing from the data

data['Price']=boston_dataset.target
data.head()
```

```python
#Displot
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.displot(data['Price'], bins=30)
plt.show()
```

```python
correlation_matrix=data.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True);
```

```python
plt.figure(figsize=(20,5))
features=['LSTAT','RM']
target=data['Price']
for i,col in enumerate(features):
    plt.subplot(1,len(features),i+1)
    x=data[col]
    y=target
    plt.scatter(x,y,marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
    
    
    
    
    
```

```python
import numpy as np
X=pd.DataFrame(np.c_[data['LSTAT'],data['RM']],columns=['LSTAT','RM'])
Y=data['Price']
```

```python
#Splitting the data into trainint and testing
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Training and testing the model
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,Y_train
```

```python
#Model Evaluation

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#model evaluation for testing set
y_pred=model.predict(X_test)

rmse=(np.sqrt(mean_squared_error(Y_test,y_pred)))
r2=r2_score(Y_test,y_pred)   
      
print("The model performance for testing set:")

print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))

```

```python
#Predicting Selling price
sample_data=[[6.89,9.939]]

#Show Predictions
price=model.predict(sample_data)
print("Predicted Selling price for house: $ {:,.2f}",format(price[0]))


```
