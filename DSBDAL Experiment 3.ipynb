```python
#For Question 1
import pandas as pd
```

```python
#load dataset
df = pd.read_csv("D:\Sem 6\DSBDA Pract\income.csv")
#df1.head()
```

```python
# Show the conetnt from dataset.
df.head(10)
```

```python
#Using groupby()
df.groupby(df.AGE_GROUP).count()
```

```python
#minimum value of income in each age group
df.groupby(df.AGE_GROUP).min()
```

```python
#maximum value of income in each age group
df.groupby(df.AGE_GROUP).max()
```

```python
#mean of income in each age group
df.groupby(df.AGE_GROUP).mean()
```

```python
#standard deviation of income in each age group
df.groupby(df.AGE_GROUP).std()
```

```python
#describe function gives count,mean,standard deviation,min,max,25th percentile,50th percentile and 75th percentile
df.groupby(df.AGE_GROUP).describe()
```

```python
#part 2 For Iris Dataset

#import libraries
from sklearn import datasets
data = datasets.load_iris()
df = pd.DataFrame(data.data,columns=data.feature_names)
df['species'] = pd.Series(data.target)
df.head()
```

```python
#Unique function to check what values it takes
df.species.unique()
```

```python
#count of all quantitative variable according to categorial variable
df.groupby(df.species).count()
```

```python
#max value of each categorial variable as species
df.groupby(df.species).max()
```

```python
#min value of each categorial variable as species
df.groupby(df.species).min()
```

```python
#mean value of each categorial variable as species

df.groupby(df.species).mean()
```

```python
#standard deviation of each categorial variable as specie
df.groupby(df.species).std()
```

```python
#describe function gives count,mean,standard deviation,min,max,25th percentile,50th percentile and 75th percentile
df.groupby(df.species)["sepal length (cm)"].describe()
```

```python
df.groupby(df.species)["sepal width (cm)"].describe()
```

```python
#statistics for petal length
df.groupby(df.species)["petal length (cm)"].describe()
```

```python
#statistics for petal width
df.groupby(df.species)["petal width (cm)"].describe()
```
